```python
import math
import os

class ThuatToanMinimaxNxN:
    def __init__(self, kich_thuoc, dieu_kien_thang, do_sau_toi_da):
        self.kich_thuoc = kich_thuoc
        self.dieu_kien_thang = dieu_kien_thang
        self.do_sau_toi_da = do_sau_toi_da

    def tim_nuoc_di_tot_nhat(self, ban_co, nguoi_choi):
        gia_tri_tot_nhat = -math.inf
        nuoc_di_tot_nhat = [-1, -1]

        for hang in range(self.kich_thuoc):
            for cot in range(self.kich_thuoc):
                if ban_co[hang][cot] == 0:
                    ban_co[hang][cot] = nguoi_choi
                    gia_tri = self.minimax(ban_co, 0, False, nguoi_choi)
                    ban_co[hang][cot] = 0

                    if gia_tri > gia_tri_tot_nhat:
                        gia_tri_tot_nhat = gia_tri
                        nuoc_di_tot_nhat = [hang, cot]
        return nuoc_di_tot_nhat

    def minimax(self, ban_co, do_sau, la_nguoi_choi_toi_da, nguoi_choi):
        doi_thu = 2 if nguoi_choi == 1 else 1

        if self.kiem_tra_thang(ban_co, nguoi_choi):
            return 1000 - do_sau
        if self.kiem_tra_thang(ban_co, doi_thu):
            return -1000 + do_sau
        if self.kiem_tra_hoa(ban_co) or do_sau == self.do_sau_toi_da:
            return self.danh_gia_ban_co(ban_co, nguoi_choi)

        if la_nguoi_choi_toi_da:
            gia_tri_toi_da = -math.inf
            for hang in range(self.kich_thuoc):
                for cot in range(self.kich_thuoc):
                    if ban_co[hang][cot] == 0:
                        ban_co[hang][cot] = nguoi_choi
                        gia_tri = self.minimax(ban_co, do_sau + 1, False, nguoi_choi)
                        ban_co[hang][cot] = 0
                        gia_tri_toi_da = max(gia_tri_toi_da, gia_tri)
            return gia_tri_toi_da
        else:
            gia_tri_toi_thieu = math.inf
            for hang in range(self.kich_thuoc):
                for cot in range(self.kich_thuoc):
                    if ban_co[hang][cot] == 0:
                        ban_co[hang][cot] = doi_thu
                        gia_tri = self.minimax(ban_co, do_sau + 1, True, nguoi_choi)
                        ban_co[hang][cot] = 0
                        gia_tri_toi_thieu = min(gia_tri_toi_thieu, gia_tri)
            return gia_tri_toi_thieu

    def kiem_tra_thang(self, ban_co, nguoi_choi):
        for hang in range(self.kich_thuoc):
            for cot in range(self.kich_thuoc):
                if ban_co[hang][cot] == nguoi_choi:
                    if (self.kiem_tra_huong(ban_co, hang, cot, 1, 0, nguoi_choi) or
                        self.kiem_tra_huong(ban_co, hang, cot, 0, 1, nguoi_choi) or
                        self.kiem_tra_huong(ban_co, hang, cot, 1, 1, nguoi_choi) or
                        self.kiem_tra_huong(ban_co, hang, cot, 1, -1, nguoi_choi)):
                        return True
        return False

    def kiem_tra_huong(self, ban_co, hang, cot, buoc_hang, buoc_cot, nguoi_choi):
        dem = 0
        for i in range(self.dieu_kien_thang):
            hang_moi = hang + i * buoc_hang
            cot_moi = cot + i * buoc_cot

            if 0 <= hang_moi < self.kich_thuoc and 0 <= cot_moi < self.kich_thuoc and ban_co[hang_moi][cot_moi] == nguoi_choi:
                dem += 1
            else:
                break
        return dem == self.dieu_kien_thang

    def kiem_tra_hoa(self, ban_co):
        for hang in ban_co:
            for o_co in hang:
                if o_co == 0:
                    return False
        return True

    def danh_gia_ban_co(self, ban_co, nguoi_choi):
        return 0

def xoa_man_hinh():
    os.system('cls' if os.name == 'nt' else 'clear')

def in_ban_co(ban_co, kich_thuoc):
    ky_hieu = {0: " . ", 1: " O ", 2: " X "}
    cot_header = "     " + "   ".join([f"{i:2}" for i in range(kich_thuoc)])
    print(cot_header)
    print("   " + "-" * (kich_thuoc * 5 + 1))
    for i in range(kich_thuoc):
        hang_in = f"{i:2} |"
        for j in range(kich_thuoc):
            hang_in += ky_hieu[ban_co[i][j]] + "|"
        print(hang_in)
        print("   " + "-" * (kich_thuoc * 5 + 1))

def choi_game():
    xoa_man_hinh()
    print("--- TRÒ CHƠI TIC-TAC-TOE VS AI (Ma trận N x N) ---")

    try:
        kich_thuoc = int(input("Nhập kích thước bàn cờ N (VD: 5 cho 5x5, 10 cho 10x10): "))
        dieu_kien_thang = int(input("Nhập số quân liên tiếp để thắng (VD: 4 hoặc 5): "))
        do_sau = int(input("Nhập độ sâu tính toán của AI (Khuyến nghị 2-3 cho bàn cờ lớn): "))
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng chạy lại chương trình và nhập số nguyên.")
        return

    ai = ThuatToanMinimaxNxN(kich_thuoc, dieu_kien_thang, do_sau)
    ban_co = [[0 for _ in range(kich_thuoc)] for _ in range(kich_thuoc)]

    nguoi_choi_hien_tai = 2

    while True:
        xoa_man_hinh()
        print(f"Bàn cờ {kich_thuoc}x{kich_thuoc} | Thắng: {dieu_kien_thang} quân | AI Độ sâu: {do_sau}")
        print("Bạn là: X | AI là: O")
        in_ban_co(ban_co, kich_thuoc)

        if ai.kiem_tra_thang(ban_co, 2):
            print("\nChúc mừng! Bạn đã thắng!")
            break
        elif ai.kiem_tra_thang(ban_co, 1):
            print("\nAI đã thắng! Chúc bạn may mắn lần sau.")
            break
        elif ai.kiem_tra_hoa(ban_co):
            print("\nTrận đấu hòa!")
            break

        if nguoi_choi_hien_tai == 2:
            try:
                print("\nĐến lượt bạn.")
                hang = int(input(f"Nhập số hàng (0-{kich_thuoc - 1}): "))
                cot = int(input(f"Nhập số cột (0-{kich_thuoc - 1}): "))

                if 0 <= hang < kich_thuoc and 0 <= cot < kich_thuoc and ban_co[hang][cot] == 0:
                    ban_co[hang][cot] = 2
                    nguoi_choi_hien_tai = 1
                else:
                    print("Nước đi không hợp lệ. Vui lòng nhấn Enter để thử lại.")
                    input()
            except ValueError:
                print("Vui lòng nhập một số hợp lệ. Nhấn Enter để thử lại.")
                input()
        else:
            print("\nAI đang suy nghĩ...")
            nuoc_di_ai = ai.tim_nuoc_di_tot_nhat(ban_co, 1)
            if nuoc_di_ai[0] != -1:
                ban_co[nuoc_di_ai[0]][nuoc_di_ai[1]] = 1
            nguoi_choi_hien_tai = 2

if __name__ == "__main__":
    choi_game()
```
